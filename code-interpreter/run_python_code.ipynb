```python
# 运行python脚本的不同方式
import pandas as pd
a = 1
# 本地运行
exec("print('hello world',a)")
```

```python
# 使用subprocess模块本地运行，但是他不会获取到当前的一些变量，会比上面的稍微安全一点点，但是依然不安全 如果用户传入恶意代码，依然会被执行
import subprocess

subprocess.run(["python", "-c", "print('hello world')"])  # 此时如果带上a会报错

subprocess.run(["python", "./temp.py"])
```

```python
result = subprocess.run(["python", "./temp.py"], capture_output=True, text=True, timeout=10)

print("stdout:", result.stdout)
```

```python
import subprocess

try:
    result = subprocess.run(
        ["python", "-c", "print1('hello')"], capture_output=True, text=True, timeout=10
    )
except Exception as e:
    print("运行代码时出错:", e)


print("stdout:", result.stdout)
print("stderr:", result.stderr)
```
